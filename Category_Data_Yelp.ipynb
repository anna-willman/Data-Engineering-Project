## De-dupe categories

```python
import pandas as pd
import numpy as np
```

```python
categories = pd.read_csv('yelp_categories_2.csv', encoding = 'ISO-8859-1', header = None)
```

```python
categories
```

```python
categories.shape
```

```python
all_categories = categories[0].append(categories[1])
```

```python
all_categories.shape
```

```python
for i in range(2,35):
    all_categories = all_categories.append(categories[i])
```

```python
all_categories.shape
```

```python
all_categories_dedupe = all_categories.drop_duplicates()
```

```python
all_categories_dedupe.shape
```

```python
all_categories_dedupe = all_categories_dedupe.dropna()
```

```python
all_categories_dedupe.shape
```

```python
isinstance(all_categories_dedupe, pd.DataFrame)
```

```python
all_categories_dedupe_df = pd.DataFrame(all_categories_dedupe)
```

```python
all_categories_dedupe_df.insert(loc=0, column='category_id', value=np.arange(len(all_categories_dedupe_df)))
```

```python
all_categories_dedupe_df = all_categories_dedupe_df.rename(columns={0: "categories"})
```

```python
all_categories_dedupe_df.shape
```

```python
all_categories_dedupe_df['categories'] = all_categories_dedupe_df.select_dtypes(['object']).apply(lambda x: x.str.strip())
```

```python
all_categories_dedupe_df.head()
```

## attach category ID to business id

```python
# convert category df to dictionary
category_id_dict = all_categories_dedupe_df.set_index('categories').to_dict()
```

```python
category_id_dict2 = list(category_id_dict.values())
```

```python
category_id_dict2
```

```python
category_id_dict2[0]['Fast Food']
```

```python
# bring in categories for each restaurant
```

```python
def find_cat_id(category):
    return(category_id_dict2[0][category])
```

```python
restaurants = pd.read_csv('restaurant_categories.csv', encoding = 'ISO-8859-1', header = 0)
```

```python
restaurants = pd.DataFrame(restaurants)
restaurants = restaurants.set_index('business_id')
```

```python
restaurants
```

```python
# stack category columns into rows
restaurants_transform = restaurants.stack().reset_index()
# drop level_1 column
restaurants_transform = restaurants_transform.drop(['level_1'], axis=1)
# clean data
restaurants_transform = restaurants_transform.rename(columns={0: "categories"})
restaurants_transform = pd.DataFrame(restaurants_transform)
restaurants_transform['categories'] = restaurants_transform['categories'].str.strip()
# add category_id
restaurants_transform['category_id'] = restaurants_transform['categories'].apply (lambda x: find_cat_id(x))

restaurants_transform.head()
```

```python
fact_table = restaurants_transform.drop(['categories'], axis=1)
```

```python
fact_table = pd.DataFrame(fact_table)
fact_table = fact_table.set_index('business_id')
```

```python
fact_table.shape
```

```python
fact_table.head(5)
```

```python
names = pd.read_csv('names_table2.csv', encoding = 'ISO-8859-1', header = None)
names = pd.DataFrame(names)
names = names.rename(columns={0: "business_id"})
names = names.rename(columns={1: "name"})
names = names.set_index('business_id')
names.head(5)
```

```python
fact_table = fact_table.join(names, on='business_id')
```
